```python
import pandas as pd
from functools import reduce
```

```python
df_main = pd.read_json('data/1094-v2-victor/main.json')
df_main = df_main.rename(columns={'generation': 'main'})

df_pr1094_js_false = pd.read_json('data/1094-v2-victor/1094-js-false.json')
df_pr1094_js_false = df_pr1094_js_false.rename(columns={'generation': 'PR 1094 JS false'})

df_pr1094_js_true = pd.read_json('data/1094-v2-victor/1094-js-true.json')
df_pr1094_js_true = df_pr1094_js_true.rename(columns={'generation': 'PR 1094 JS true'})
```

```python
dataframes = [df_pr1094_js_true, df_pr1094_js_false, df_main]
merged_df = reduce(lambda left, right: pd.merge(left, right, on=['prompt', 'answer']), dataframes)
```
