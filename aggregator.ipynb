```python
import pandas as pd
from functools import reduce
```

```python
df_main = pd.read_json('data/liam-collab2/main.json')
df_main = df_main.rename(columns={'generation': 'main'})

df_pr1094 = pd.read_json('data/liam-collab2/pr1094.json')
df_pr1094 = df_pr1094.rename(columns={'generation': 'PR 1094'})
```

```python
dataframes = [df_pr1094, df_main]
merged_df = reduce(lambda left, right: pd.merge(left, right, on=['prompt', 'answer']), dataframes)
```

```python
merged_df.to_csv("pr1094-2.csv")
```
